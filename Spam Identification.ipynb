```python
df = pd.read_csv('spam.csv')
# Display the first few rows of the DataFrame
df.head()
```

```python
df.dropna(subset=['v1', 'v2'], inplace=True)
#dropna(subset=['v1', 'v2'], inplace=True): This line removes any rows from the DataFrame where either the 'v1' or 'v2' column has missing (NaN) values.
df.head()
```

```python
df['label'] = df['v1'].map({'ham': 0, 'spam': 1})
#df['v1'].map({'ham': 0, 'spam': 1}): This line creates a new column 'label' based on the values in the 'v1' column.
df
```

```python
# Drop unnecessary columns
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis='columns', inplace=True)
df
```

```python
x = df['v2'] #X = df['v2']: This line assigns the 'v2' column (messages) to the variable X. X will be the input features for the model.
y = df['label'] #y = df['label']: This line assigns the 'label' column (0 for 'ham', 1 for 'spam') to the variable y. y will be the target variable or labels for the model.
x,y
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets (e.g., 70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#"CountVectorizer"transforms a set of text data into a numerical format that machine learning models can understand.

vectorizer = CountVectorizer()
#vectorizer = CountVectorizer(): This line creates an instance of the CountVectorizer class. This instance will be used to convert text data into a numerical format.

X_train_vectorized = vectorizer.fit_transform(X_train) #converts text data (messages) into numerical format
X_test_vectorized = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

# Create a Naive Bayes classifier
model = MultinomialNB()

# Train the model
model.fit(X_train_vectorized, y_train)
```
